```python
import pandas as pd
import re
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel
from sklearn.metrics import mean_squared_error

# Read data from CSV
df = pd.read_csv('coursea_data.csv')

# Preprocessing: Convert 'course_students_enrolled' to numeric
def convert_students_enrolled(value):
    if 'k' in value:
        return float(re.sub(r'[^0-9.]', '', value)) * 1000
    elif 'm' in value:
        return float(re.sub(r'[^0-9.]', '', value)) * 1000000
    else:
        return float(value)

df['course_students_enrolled'] = df['course_students_enrolled'].apply(convert_students_enrolled)

# Feature extraction
tfidf_vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf_vectorizer.fit_transform(df['course_title'] + ' ' + df['course_organization'] + ' ' + df['course_difficulty'])

# Compute the cosine similarity matrix
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)

# Combine rating and enrollment information
df['combined_score'] = df['course_rating'] * df['course_students_enrolled']

# Function to get course recommendations based on combined score
def get_recommendations(keyword, cosine_sim_matrix, df):
    keyword = keyword.strip().lower()  # Remove leading/trailing whitespaces and convert to lowercase
    matches = df['course_title'].apply(lambda title: keyword in title.lower())
    matched_indices = df[matches].index

    if len(matched_indices) == 0:
        print(f"No courses found matching the keyword '{keyword}'.")
        return pd.DataFrame()

    # Calculate combined score for matched courses
    combined_scores = df.loc[matched_indices, 'combined_score'].tolist()

    # Get top 3 courses based on combined score
    top_courses = sorted(zip(matched_indices, combined_scores), key=lambda x: x[1], reverse=True)[:3]
    top_indices = [idx for idx, _ in top_courses]

    return df[['course_title', 'course_rating', 'course_students_enrolled', 'combined_score']].iloc[top_indices]

# Example: Get recommendations for a specific keyword
keyword = 'python'
recommendations = get_recommendations(keyword, cosine_sim, df)
print(recommendations)

# Visualization: Distribution of Ratings
plt.figure(figsize=(10, 6))
sns.histplot(df['course_rating'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Course Ratings')
plt.xlabel('Course Rating')
plt.ylabel('Frequency')
plt.show()

# Visualization: Distribution of Enrolled Students
plt.figure(figsize=(10, 6))
sns.histplot(df['course_students_enrolled'], bins=20, kde=True, color='salmon')
plt.title('Distribution of Enrolled Students')
plt.xlabel('Number of Enrolled Students')
plt.ylabel('Frequency')
plt.show()

# Calculate Mean Squared Error (MSE)
actual_ratings = df['course_rating']
actual_students_enrolled = df['course_students_enrolled']
predicted_ratings = df['combined_score'] / df['course_students_enrolled']  # Assuming normalized rating per student

mse_ratings = mean_squared_error(actual_ratings, predicted_ratings)
mse_students_enrolled = mean_squared_error(actual_students_enrolled, df['combined_score'])

print(f'Mean Squared Error (MSE) for Ratings: {mse_ratings:.4f}')
print(f'Mean Squared Error (MSE) for Enrolled Students: {mse_students_enrolled:.4f}')

# Report Generation
## Visualization 1: Bar chart of the top recommended courses
plt.figure(figsize=(10, 6))
sns.barplot(x='combined_score', y='course_title', data=recommendations, palette='viridis')
plt.title(f'Top Recommended Courses related to "{keyword}"')
plt.xlabel('Combined Score')
plt.ylabel('Course Title')
plt.show()

## Visualization 2: Scatter plot of Rating vs Enrolled Students
plt.figure(figsize=(10, 6))
sns.scatterplot(x='course_rating', y='course_students_enrolled', data=df, hue='course_difficulty', palette='Set2', size='combined_score', sizes=(50, 200))
plt.title('Scatter Plot of Rating vs Enrolled Students with Difficulty as Hue')
plt.xlabel('Course Rating')
plt.ylabel('Course Students Enrolled')
plt.legend(title='Difficulty')
plt.show()
```
